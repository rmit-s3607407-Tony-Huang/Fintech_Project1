```python
import yfinance as yf
import pandas as pd
from datetime import datetime
```

```python
GSPC = yf.Ticker("^GSPC")
print(GSPC)
```

```python
GSPC.info
```

```python
SP500 = GSPC.history(interval="1d", start="2016-01-01", end="2021-09-28")
SP500
```

```python
SP500 = SP500.drop('Stock Splits', axis=1)
```

```python
SP500 = SP500.drop('Volume', axis=1)
```

```python
SP500 = SP500.drop(SP500.index[0])
```

```python
pd.to_datetime(SP500.index, unit='s')
SP500
```

```python
SP500.index = SP500.index.strftime('%Y-%m-%d %H:%M:%S')
```

```python
SP500
```
