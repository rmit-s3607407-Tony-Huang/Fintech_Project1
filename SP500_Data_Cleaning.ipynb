```python
import yfinance as yf
```

```python
GSPC = yf.Ticker("^GSPC")
print(GSPC)
```

```python
GSPC.info
```

```python
SP500 = GSPC.history(interval="1d", start="2016-01-01", end="2021-09-28")
SP500
```

```python
SP500 = SP500.drop('Stock Splits', axis=1)
```

```python
SP500 = SP500.drop('Dividends', axis=1)
```

```python
SP500 = SP500.drop('Volume', axis=1)
```

```python
SP500 = SP500.drop('2015-12-31', axis=0)
```

```python
SP500.to_csv(r'C:\Users\Ashley\Desktop\SP500_Data.csv')
```
